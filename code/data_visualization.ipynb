```python
import pandas as pd
from preprocessing import DataframeCreator, Spellchecker, Preprocessor
```

```python
# Data
data_file = open("../data/raw/training-v1.1/de/HIPE-data-v1.1-train-de.tsv", "r+")

# List of dataframes for each newspaper
dfs = DataframeCreator().create_dataframes(data_file)

# Joint dataframe
df = pd.concat(dfs, ignore_index=True)
```
